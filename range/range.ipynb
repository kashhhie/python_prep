```python
#wpp to print count of even and odd no. between 1000 and 2000
even_count=0
odd_count=0
for i in range(1000,2001):
    if i%2==0:
        even_count =even_count+1
    else:
        odd_count =odd_count+1
print(odd_count)
print(even_count)
```

```python
#wpp to print factorial of n number
n=int(input("any number: "))
factorial=1
for i in range(1,n+1):
    factorial=factorial*i
print(factorial)
```
