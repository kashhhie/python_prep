```python
# numeric
#integer
a=10
print(a)
print(type(a))
```

```python
a=10.0
print(a)
print(type(a))
```

```python
#string - ordered sequence of character
#single line denoted by '' or "" and
#multiline string denoted ''' '''
#indexed and immutable
s="" #empty string
print(s)
print(type(s))
```

```python
s="decode"
print(s)
print(type(s))
print(s[0])
s[0]="z"
print(s)
```

```python
s='''hello
welcome
to bhopal.'''
print(s)
```

```python
#list - collection of element
#it is denoted by [] with csv
#indexed,ordered,mutable and heterogeneous in nature
l=[]
print(l)
print(type(l))
```

```python
l=[10,20,30,'decode',10.5]
print(l)
print(l[3])
l[0]="bhopal"
print(l)
```

```python
#tuple collection of elements
#denoted by () with csv
#ordered,indexed,immutable and heterogeneous in nature
t=()
print(t)
print(type(t))
```

```python
#single value tupple
t=('10.5',)
print(t)
print(type(t))
```

```python
t=(10,20,300,"dcode",10.5)
print(t)
print(t[2])
t[2]=400
print(t)
```

```python
#dictionary
#collection datavalue
#it is denoted by (k:v) with key:value pair
#ordered,unindexed, and mutable.
#keys---unique
#values===duplicate.
d={}
print(d)
print(type(d))
```

```python
d={"name":"abc","subject":"da","fees":"30000"}
print(d)
print(d["name"])
d["name"]="def" 
print(d)
```

```python
d={}
d["name"]="def"
print(d)
```

```python
d={"name":["def","abc","pqr"],"subject":["da","da","ds"]}
print(d)
print(d["name"])
```

```python
#set - collection of element
#it is denoted by {} with csv
#unordered,unindexed,mutable,heterogenous in nature
#store unique value.
s=set()
print(s)
print(type(s))
```

```python
s={10,10,10,20,10,100,200,"decode","h"}
print(s)
print(type(s))
s.add(1000)
print(s)      
```
