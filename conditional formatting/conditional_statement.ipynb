```python
#control flow statement
#conditional formatting
#if
num=10
if num%2==0:
    print(num, "is an even number")
```

```python
#if else
num=int(input("enter any number: "))
if num%2==0:
    print(num, "is an even number")
else:
    print(num, "is an odd number")
```
